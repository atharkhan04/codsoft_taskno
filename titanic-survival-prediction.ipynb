```python
from sklearn.model_selection import train_test_split

# Function to split data into train and test sets
def splitting_train_test(x, y):
    x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42, shuffle=True)
    print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)  # Print shapes
    return x_train, x_test, y_train, y_test  # Return the values

x_train, x_test, y_train, y_test = splitting_train_test(x_data, y_data)  
  
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
log_reg=LogisticRegression()
```

```python
def all(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    accuracy_test=accuracy_score(y_pred,y_test)*100
    accuracy_train=model.score(x_train,y_train)*100
    print('accuracy after train model :',accuracy_train)
    print('accuracy after test  model :',accuracy_test)
```

```python
all(log_reg,x_train, x_test, y_train, y_test)
```
